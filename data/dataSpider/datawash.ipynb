```python
import pandas as pd
```

```python
filename = 'song_list_unduplicated.csv'
names = ['albumname','songname','singer','songmid','songid','tags','visitnum']
data = pd.read_csv(filename, names = names, encoding='utf-8')
```

```python
data.drop([0], inplace = True)
data.reset_index(drop=True)
```

```python
df1 = data['singer'].str.split(',', expand = True)
data2 = pd.merge(data, df1, how = 'left', left_index=True, right_index=True)
df2 = data['tags'].str.split(',', expand = True)
data = pd.merge(data2, df2, how = 'left', left_index=True, right_index=True)
data = data.drop(['singer','tags'], axis = 1)
```
