```python
import pandas as pd
```

```python
filename = 'song_unduplicated.csv'
names = ['albumname','songname','singer','singermid','songmid','tags','tagsid','visitnum','albummid','url']
data = pd.read_csv(filename, names = names, encoding='utf-8')
```

```python
data.drop([0], inplace = True)
data.reset_index(drop=True)
```

```python
df1 = data[['songmid','singermid']]
df1.columns = ['songmid','singermid']
df2 = df1['singermid'].str.split(',', expand = True).stack()
df2 = df2.reset_index(level=1,drop=True)
df2.name = 'singermid'
df_new = df1.drop(['singermid'], axis = 1).join(df2)
df_new.reset_index(drop=True)
df_new.to_csv('song_singer.csv',encoding = 'utf-8', index=False)
```

```python
df1 = data[['songmid', 'tagsid']]
df1.columns = ['songmid', 'tagsid']
df2 = df1['tagsid'].str.split(',', expand = True).stack()
df2 = df2.reset_index(level=1,drop=True)
df2.name = 'tagsid'
df_new = df1.drop(['tagsid'], axis = 1).join(df2)
df_new.reset_index(drop=True)
df_new.to_csv('song_tag.csv',encoding = 'utf-8', index=False)
```

```python
df1 = data[['songmid','albummid']]
df1.columns = ['songmid','albummid']
df1.to_csv('song_album.csv',encoding = 'utf-8', index=False)
```
