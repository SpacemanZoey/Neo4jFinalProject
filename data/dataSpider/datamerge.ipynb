```python
import pandas as pd
import re
```

```python
filename = 'song_list.csv'
names = ['number', 'albumname','songname','singer','songmid','songid','tags','visitnum']
data = pd.read_csv(filename, names = names, encoding='utf-8')
data
```

```python
data.drop([0], inplace = True)
data.drop(['number'], axis=1, inplace = True)
data.reset_index(drop=True)
```

```python
ndf=pd.DataFrame(columns=data.columns)
uids=set(data['songid'])
```

```python
for u in uids:
    one=data.loc[data['songid']==u]
    taglst=list(one['tags'])
    visitlst=list(one['visitnum'])
    zero=one.iloc[[0]]
    if len(taglst) > 1:
        tag = taglst[0]
        tag = filter(None, re.split(r',| ', tag))
        for t in taglst:
            tag = list(set(tag))
            t = filter(None, re.split(r',| ', t))
            for x in t:
                if tag.count(x) == 0:
                    tag.append(x)
        zero['tags'] = ','.join(tag)
        visitnum = 0
        for v in visitlst:
            visitnum += int(v)
        zero['visitnum'] = visitnum
    ndf=pd.concat([ndf,zero])
ndf
```

```python
ndf.reset_index(drop = True)
ndf.to_csv('song_list_unduplicated.csv',index=False)
```
