```python
import pandas as pd
import re
```

```python
filename = 'song.csv'
names = ['albumname','songname','singer','singermid','songmid','tags','tagsid','visitnum','albummid','url']
data = pd.read_csv(filename, names = names, encoding='utf-8')
data
```

```python
data.drop([0], inplace = True)
data.reset_index(drop=True)
```

```python
ndf=pd.DataFrame(columns=data.columns)
uids=set(data['songmid'])
```

```python
for u in uids:
    one=data.loc[data['songmid']==u]
    taglst=list(one['tags'])
    visitlst=list(one['visitnum'])
    tagidlst = list(one['tagsid'])
    zero=one.iloc[[0]]
    if len(taglst) > 1:
        tag = taglst[0]
        tag = filter(None, re.split(r',| ', tag))
        tagid = tagidlst[0]
        tagid = filter(None, re.split(r',| ', tagid))
        for t in taglst:
            tag = list(set(tag))
            t = filter(None, re.split(r',| ', t))
            for x in t:
                if tag.count(x) == 0:
                    tag.append(x)
        for ti in tagidlst:
            tagid = list(set(tagid))
            ti =  filter(None, re.split(r',| ', ti))
            for x in ti:
                if tagid.count(x)==0:
                    tagid.append(x)
        zero['tags'] = ','.join(tag)
        zero['tagsid'] = ','.join(tagid)
        visitnum = 0
        for v in visitlst:
            visitnum += int(v)
        zero['visitnum'] = visitnum
    ndf=pd.concat([ndf,zero])
ndf
```

```python
ndf.reset_index(drop = True)
```

```python
ndf.dropna(axis=0, how='any', inplace=True)
ndf
```

```python
ndf.to_csv('song_unduplicated.csv',index=False)
```
